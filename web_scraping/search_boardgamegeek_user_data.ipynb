```python
import time
import selenium.webdriver
from bs4 import BeautifulSoup
import itertools
import pandas as pd
import re
from selenium.common.exceptions import NoSuchElementException
```

```python
def search_boardgamegeek(query, browser):
    search_url = "https://boardgamegeek.com/findgamers.php?action=findclosest&country=US&srczip={}&maxdist=100&B1=Submit".format(query)
    browser.get(search_url)
    return browser.page_source

def get_boardgamegeek_usernames(query, browser):
    html = search_boardgamegeek(query, browser)
    soup = BeautifulSoup(html, 'html.parser')
    product_tags = soup.find_all("div", class_="username")
    product_tags.extend(soup.find_all("div", class_="username"))
    return [get_user_url(t) for t in product_tags]

def get_user_url(product_tag):
    title_elements = product_tag.select('a')
    if title_elements:
        product_title = title_elements[0].get('href')
        return product_title
    
browser = selenium.webdriver.Firefox()
for i in xrange(17601, 99951, 20):
    zip_ = "{0:0=5d}".format(i)
    username_urls.append(get_boardgamegeek_usernames(zip_, browser))
    us_users_urls = list(itertools.chain.from_iterable(username_urls))
    us_users_urls = list(set(us_users_urls))
browser.quit()

us_users_urls = list(itertools.chain.from_iterable(username_urls))
us_users_urls = list(set(us_users_urls))
us_users = []
for user in us_users_urls:
    us_users.append(user.split('/')[2])
```
